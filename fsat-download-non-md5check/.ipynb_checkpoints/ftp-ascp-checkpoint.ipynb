```python
f = open('263122.txt','r')
f1 = open('download_263122.sh','w')
lines = f.readlines()
ex_line = 'ascp -QT -l 300m -P33001 -i /Users/mac/Applications/Aspera\ CLI/etc/asperaweb_id_dsa.openssh era-fasp@fasp.sra.ebi.ac.uk:'
back_line =' /Users/mac/Desktop/1'
scp_ex = 'scp /Users/mac/Desktop/1/'
scp_back = ' will@10.75.119.200:/home/will/data_store/bird-262'
for line in lines[260:263]:
    line = line.strip()
    elements = line.split()
    sample = elements[1][3:]
    acps = elements[2].split(';')
    acp_1 = acps[0][17:]
    acp_2 = acps[1][17:]
    d1 = ex_line+acp_1+back_line
    d2 = ex_line+acp_2+back_line
    t1 = scp_ex+acp_1[35:]+scp_back
    t2 = scp_ex+acp_2[35:]+scp_back
    r1 = 'rm '+t1[4:45]
    r2 = 'rm '+t2[4:45]
    
    print(d1)
    f1.write(d1+'\n')
    print(t1)
    f1.write(t1+'\n')
    print(r1)
    f1.write(r1+'\n')
    
    print(d2)
    f1.write(d2+'\n')
    print(t2)
    f1.write(t2+'\n')
    print(r2)
    f1.write(r2+'\n')
    
f.close()
f1.close()
```
